```python
# Numpy, Matplotlib ãŠã‚ˆã³ æ—¥æœ¬èªåŒ–ãƒ©ã‚¤ãƒ–ãƒ©ãƒªãƒ¼ã®èª­ã¿è¾¼ã¿
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

# -----------------------------------------------------------------
# æ¸›è¡°æŒ¯å‹•ï¼ˆä¸è¶³æ¸›è¡°ï¼‰ã®ç†è«–è§£ã‚’è¨ˆç®—ã™ã‚‹é–¢æ•° (ãƒ™ã‚¯ãƒˆãƒ«åŒ–)
# u(t) = exp(-h*wn*t) * (C1*cos(wd*t) + C2*sin(wd*t))
# -----------------------------------------------------------------
def calc_underdamped_response(t, wn, h, u0, v0):
    """
    ä¸è¶³æ¸›è¡°(h < 1)ã®å¤‰ä½å¿œç­”u(t)ã‚’Numpyãƒ™ã‚¯ãƒˆãƒ«åŒ–ã§è¨ˆç®—ã™ã‚‹
    """
    # æ¸›è¡°å›ºæœ‰å††æŒ¯å‹•æ•° Ï‰d
    wd = wn * np.sqrt(1.0 - h**2)

    # ç©åˆ†å®šæ•° C1, C2
    C1 = u0
    C2 = (v0 + h * wn * u0) / wd

    # ç†è«–è§£ã®è¨ˆç®— (Numpyé…åˆ—tå…¨ä½“ã«ä¸€åº¦ã«é©ç”¨)
    u = np.exp(-h * wn * t) * (C1 * np.cos(wd * t) + C2 * np.sin(wd * t))
    return u

# =================================================================
# æ¼”ç¿’èª²é¡Œ 1
# h=0.05, 5ç§’é–“, 4ã‚±ãƒ¼ã‚¹ã®æ¯”è¼ƒ
# =================================================================
print("æ¼”ç¿’èª²é¡Œ1ã®ã‚°ãƒ©ãƒ•ã‚’ç”Ÿæˆä¸­...")

# --- 1.1. è¨ˆç®—æ¡ä»¶ ---
h = 0.05
t1 = np.linspace(0, 5, 501) # 0ã‹ã‚‰5ç§’ã¾ã§ (0.01ç§’åˆ»ã¿)

# èª²é¡Œ1ã®4ã‚±ãƒ¼ã‚¹ã®ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿
cases_1 = {
    'Case A': {'wn': np.pi / 2, 'u0': 0.0, 'v0': -10.0},
    'Case B': {'wn': np.pi / 2, 'u0': 5.0, 'v0': -20.0},
    'Case C': {'wn': np.pi,     'u0': 0.0, 'v0': -10.0},
    'Case D': {'wn': np.pi,     'u0': 5.0, 'v0': -20.0},
}

# --- 1.2. ã‚°ãƒ©ãƒ•æç”» (2x2 ã®ã‚µãƒ–ãƒ—ãƒ­ãƒƒãƒˆ) ---
# fig1: å›³å…¨ä½“, axes1: 4ã¤ã®ã‚°ãƒ©ãƒ•ï¼ˆaxï¼‰ã‚’å…¥ã‚Œã‚‹é…åˆ—
fig1, axes1 = plt.subplots(2, 2, figsize=(14, 10), tight_layout=True)
fig1.suptitle(f'æ¼”ç¿’èª²é¡Œ1: æ¸›è¡°è‡ªç”±æŒ¯å‹•å¿œç­” (æ¸›è¡°æ¯” h = {h})', fontsize=16, y=1.03)

# axes1.flat ã§ 2x2 ã® ax ã‚’ 1æ¬¡å…ƒã«å±•é–‹ã—ã€zipã§ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã¨çµ„ã¿åˆã‚ã›ã‚‹
for ax, (case_name, params) in zip(axes1.flat, cases_1.items()):
    wn = params['wn']
    u0 = params['u0']
    v0 = params['v0']

    # é–¢æ•°ã‚’å‘¼ã³å‡ºã—ã¦ u(t) ã‚’è¨ˆç®—
    u = calc_underdamped_response(t1, wn, h, u0, v0)

    # è©²å½“ã™ã‚‹ã‚µãƒ–ãƒ—ãƒ­ãƒƒãƒˆ(ax)ã«æç”»
    ax.plot(t1, u)
    ax.set_title(f'{case_name} (Ï‰n={wn:.2f}, u0={u0}, v0={v0})')
    ax.set_xlabel('æ™‚é–“ t (s)')
    ax.set_ylabel('å¤‰ä½ u(t)')
    ax.grid(True)
    ax.axhline(0, color='black', linewidth=0.5)

plt.show()

# =================================================================
# æ¼”ç¿’èª²é¡Œ 2
# Ï‰n=2Ï€, u0=0, v0=-10, 10ç§’é–“, hã‚’4é€šã‚Šå¤‰ãˆã‚‹
# =================================================================
print("æ¼”ç¿’èª²é¡Œ2ã®ã‚°ãƒ©ãƒ•ã‚’ç”Ÿæˆä¸­...")

# --- 2.1. è¨ˆç®—æ¡ä»¶ ---
wn = 2.0 * np.pi
u0 = 0.0
v0 = -10.0
h_values = [0.1, 0.05, 0.02, 0.01] # èª²é¡Œã§æŒ‡å®šã•ã‚ŒãŸ4ã‚±ãƒ¼ã‚¹
t2 = np.linspace(0, 10, 1001) # 0ã‹ã‚‰10ç§’ã¾ã§ (0.01ç§’åˆ»ã¿)

# --- 2.2. ã‚°ãƒ©ãƒ•æç”» (1ã¤ã®ã‚°ãƒ©ãƒ•ã«é‡ã­æã) ---
plt.figure(figsize=(12, 7))

for h in h_values:
    # é–¢æ•°ã‚’å‘¼ã³å‡ºã—ã¦ u(t) ã‚’è¨ˆç®—
    u = calc_underdamped_response(t2, wn, h, u0, v0)

    # ã‚°ãƒ©ãƒ•ã«ãƒ—ãƒ­ãƒƒãƒˆ
    plt.plot(t2, u, label=f'h = {h}')

# --- 2.3. ã‚°ãƒ©ãƒ•ä»•ä¸Šã’ ---
plt.title(f'æ¼”ç¿’èª²é¡Œ2: æ¸›è¡°æ¯”hã®å½±éŸ¿ (Ï‰n={wn:.2f}, u0={u0}, v0={v0})', fontsize=16)
plt.xlabel('æ™‚é–“ t (s)')
plt.ylabel('å¤‰ä½ u(t)')
plt.legend(title='æ¸›è¡°æ¯” $h$') # å‡¡ä¾‹ã‚’è¡¨ç¤º
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.show()

print("ã‚°ãƒ©ãƒ•ã®ç”ŸæˆãŒå®Œäº†ã—ã¾ã—ãŸã€‚")
```

## **2. è¨­è¨ˆå›³ï¼ˆç†è«–å¼ã‚’ã€Œé–¢æ•°ã€ã«ã™ã‚‹ï¼‰**
* def calc_underdamped_response(...)
  *  def ã¯ï¼Œã‚ªãƒªã‚¸ãƒŠãƒ«ã®å‘½ä»¤ï¼ˆ**é–¢æ•°**ï¼‰ã‚’ä½œã‚‹ãŸã‚ã®ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰ã§ã™ï¼
  *  ã“ã“ã§ã¯ã€Œä¸è¶³æ¸›è¡°ï¼ˆ$h<1$ï¼‰ã®ã¨ãã®å¤‰ä½ $u(t)$ ã‚’è¨ˆç®—ã™ã‚‹ã€ã¨ã„ã†å‡¦ç†ã‚’ï¼Œcalc_underdamped_response ã¨ã„ã†åå‰ã®ã€Œè¨­è¨ˆå›³ã€ã¨ã—ã¦å®šç¾©ã—ã¦ã„ã¾ã™ï¼
* (t, wn, h, u0, v0)
  *  ã“ã®é–¢æ•°ã‚’ä½¿ã†ãŸã‚ã«å¿…è¦ãªã€Œ**å¼•æ•°**ã€ã§ã™ï¼
* wd = ..., C1 = ..., C2 = ...
  *  è¬›ç¾©ã§å­¦ã‚“ã ç†è«–å¼ï¼ˆ$\omega_d = \omega_n \sqrt{1-h^2}$ ãªã©ï¼‰ã‚’ï¼Œãã®ã¾ã¾Pythonã®ã‚³ãƒ¼ãƒ‰ã«æ›¸ãç›´ã—ã¦ã„ã¾ã™ï¼
* u = np.exp(...) * (C1 * np.cos(...) + ...)
  *  ã“ã“ãŒ**ä¸€ç•ªé‡è¦**ãªãƒã‚¤ãƒ³ãƒˆã§ã™ï¼è¬›ç¾©ãƒãƒ¼ãƒˆï¼ˆLesson_4.ipynbï¼‰ã®ã‚¹ãƒ†ãƒƒãƒ—5ã§å­¦ã‚“ã **ã€Œ**ãƒ™ã‚¯ãƒˆãƒ«åŒ–**ï¼ˆVectorizationï¼‰ã€** ã‚’ä½¿ã£ã¦ã„ã¾ã™ï¼
  *  å¼•æ•° t ã¯ï¼Œnp.linspace() ã§ä½œã‚‰ã‚ŒãŸã€Œæ™‚é–“ã®é…åˆ—ã€ã§ã™ï¼
  *  np.exp(t) ã‚„ np.cos(t) ã¨æ›¸ãã ã‘ã§ï¼Œt ã®ä¸­èº«ï¼ˆä¾‹ï¼št=0.01, 0.02, ...ï¼‰ã®**ã™ã¹ã¦ã«å¯¾ã—ã¦ä¸€åº¦ã«**æŒ‡æ•°é–¢æ•°ã‚„cosã‚’è¨ˆç®—ã—ã¦ãã‚Œã¾ã™ï¼
* return u
  *  è¨ˆç®—ã—ã¦æ±‚ã‚ãŸ $u(t)$ ã®é…åˆ—ã‚’ï¼Œã“ã®é–¢æ•°ã®ã€Œ**å®Œæˆå“ï¼ˆæˆ»ã‚Šå€¤**ï¼‰ã€ã¨ã—ã¦è¿”ã™ï¼Œã¨ã„ã†æ„å‘³ã§ã™ï¼


```python
# -----------------------------------------------------------------
# æ¸›è¡°æŒ¯å‹•ï¼ˆä¸è¶³æ¸›è¡°ï¼‰ã®ç†è«–è§£ã‚’è¨ˆç®—ã™ã‚‹é–¢æ•° (ãƒ™ã‚¯ãƒˆãƒ«åŒ–)
# u(t) = exp(-h*wn*t) * (C1*cos(wd*t) + C2*sin(wd*t))
# -----------------------------------------------------------------
def calc_underdamped_response(t, wn, h, u0, v0):
    """
    ä¸è¶³æ¸›è¡°(h < 1)ã®å¤‰ä½å¿œç­”u(t)ã‚’Numpyãƒ™ã‚¯ãƒˆãƒ«åŒ–ã§è¨ˆç®—ã™ã‚‹
    """
    # æ¸›è¡°å›ºæœ‰å††æŒ¯å‹•æ•° Ï‰d
    wd = wn * np.sqrt(1.0 - h**2)

    # ç©åˆ†å®šæ•° C1, C2
    C1 = u0
    C2 = (v0 + h * wn * u0) / wd

    # ç†è«–è§£ã®è¨ˆç®— (Numpyé…åˆ—tå…¨ä½“ã«ä¸€åº¦ã«é©ç”¨)
    u = np.exp(-h * wn * t) * (C1 * np.cos(wd * t) + C2 * np.sin(wd * t))
    return u
```
